```python
corona_confirmed = corona_confirmed.drop(columns=['Lat', 'Long'])
corona_confirmed['Case_Type'] = 'Confirmed'

corona_deaths = corona_deaths.drop(columns=['Lat', 'Long'])
corona_deaths['Case_Type'] = 'Deaths'

corona_recovered = corona_recovered.drop(columns=['Lat', 'Long'])
corona_recovered['Case_Type'] = 'Recovered'
```

```python
corona = pd.concat([corona_confirmed, corona_deaths, corona_recovered])

cols = list(corona)
cols.insert(1, cols.pop(cols.index('Case_Type')))
corona = corona.loc[:, cols]
corona
```

```python
test = corona.groupby(["Country/Region", "Case_Type"]).sum()
#     ["Country/Region"].max().sort_values(ascending=False)
# test.max()
test

# idx = df.groupby(level=0)[].max().sort_values(ascending=False).index
```

```python
# Multi-indexing
# https://www.youtube.com/watch?v=tcRGa2soc-c

country_confirmed_total = corona_confirmed.groupby("Country/Region").sum().sort_values(by=corona_confirmed.columns[-1], ascending=False)


country_deaths_total = corona_deaths.groupby("Country/Region").sum().sort_values(by=corona_deaths.columns[-1], ascending=False)

country_recovered_total = corona_recovered.groupby("Country/Region").sum().sort_values(by=corona_recovered.columns[-1], ascending=False)

country_confirmed_total
```

```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)

country_total = corona_confirmed.groupby("Country/Region").sum().sort_values(by=corona_confirmed.columns[-1], ascending=False)

country_total
```

```python
china_total = country_total.loc['Mainland China']

china_infection_rate = []

for i in range(0, len(china_total)-1):
    china_infection_rate.append(china_total.iloc[i + 1] - china_total.iloc[i])

plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.yticks(np.arange(0, math.ceil(max(china_infection_rate) + 1000), 1000))
plt.grid(b=True, which='major', color='#d3d3d3', linestyle='-')
plt.plot(country_total.columns[1:], china_infection_rate)
plt.show()

# plt.yscale('log')
```

```python
plt.hist(china_infection_rate, bins=16)
np.mean(china_infection_rate)
```

```python
sns.distplot(china_infection_rate)
```
